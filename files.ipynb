```python
df = pd.read_csv('/content/drive/MyDrive/final_data_ISB.csv')
df['np_target_text_ada_embedding'] = df['target_text_ada_embedding'].apply(eval).apply(np.array)
df['np_response_text_ada_embedding'] = df['response_text_ada_embedding'].apply(eval).apply(np.array)
```

```python
node_2_text_embedding = {}

for index, row in df.iterrows():
    node_response = row['response_node']
    node_response_feature = row['np_response_text_ada_embedding']
    node_target = row['target_node']
    node_target_feature = row['np_target_text_ada_embedding']
    node_2_text_embedding[str(node_response)] = node_response_feature
    node_2_text_embedding[str(node_target)] = node_target_feature
```
