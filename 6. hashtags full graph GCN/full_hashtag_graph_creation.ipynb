```python
hashtag_edges = []
for index1, row1 in df_hashtags.iterrows:
    for index2, row2 in df_hashtags.iterrows:
        if(row1['hashtag']==row2['hashtag'] and row1['user']!=row2['user']):
            hashtag_edges.append((row1['user'], row2['user']))
```

```python
sources = []
targets = []

for edge in hashtag_edges:
    sources.append(edge[0])
    targets.append(edge[1])

data = {'source': sources, 'target': targets}

df_edges = pd.DataFrame(data)
```

```python
df_edges.to_csv('/content/drive/MyDrive/edges_hashtag_full_df.csv', index = False)
```
